```python
#IMPORT DEPENDENCIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
qs_df = pd.read_csv('qs_ranking.csv')
arwu_df = pd.read_csv('arwu_ranking.csv')
the_df = pd.read_csv('times_higher_education.csv')
```

```python
qs_df
```

```python
qs_df.isna().sum()
```

```python
qs_df.dropna(inplace = True)
qs_df.isna().sum()
```

```python
arwu_df
```

```python
arwu_df.isna().sum()
```

```python
qs_the_df = pd.merge(qs_df, the_df, on = 'Institution Name', how = 'outer')
```

```python
qs_the_df
```

```python
qs_the_arwu_df = pd.merge(qs_the_df, arwu_df, on= 'Institution Name', how = 'outer')
```

```python
qs_the_arwu_df
```

```python
pd.options.display.max_rows = 4000
```

```python
qs_the_arwu_df
```

```python
qs_the_arwu_df['times_higher_education_ranking'].fillna(value= 0, inplace = True)
qs_the_arwu_df['arwu_ranking'].fillna(value= 0, inplace = True)
```

```python
qs_the_arwu_df
```

```python
qs_the_df_inner_merge = pd.merge(qs_df, the_df, on = 'Institution Name', how = 'inner')
```

```python
qs_the_arwu_df_inner_merge = pd.merge(qs_the_df_inner_merge, arwu_df, on = 'Institution Name', how = 'inner')
```

```python
qs_the_arwu_df_inner_merge
```

```python
qs_the_arwu_df_inner_merge['qs_ranking_float'] = qs_the_arwu_df_inner_merge['qs_ranking'].astype(float)
```

```python
qs_the_arwu_df_inner_merge['the_ranking_float'] = qs_the_arwu_df_inner_merge['times_higher_education_ranking'].astype(float)
```

```python
qs_the_arwu_df_inner_merge['arwu_ranking_float'] = qs_the_arwu_df_inner_merge['arwu_ranking'].astype(float)
```

```python
qs_the_arwu_df_inner_merge['Average Ranking'] = (qs_the_arwu_df_inner_merge['qs_ranking_float'] + qs_the_arwu_df_inner_merge['the_ranking_float'] + qs_the_arwu_df_inner_merge['arwu_ranking_float'])/3
```

```python
qs_the_arwu_df_inner_merge
```

```python
qs_the_arwu_df_inner_merge.drop(axis = 1, columns = ['qs_ranking', 'times_higher_education_ranking', 'arwu_ranking'], inplace = True)
```

```python
qs_the_arwu_df_inner_merge.sort_values('Average Ranking', ascending = True, inplace = True)
```

```python
qs_the_arwu_df_inner_merge = qs_the_arwu_df_inner_merge.reset_index(drop = True)
```

```python
qs_the_arwu_df_inner_merge
```

```python
qs_the_arwu_df_inner_merge.index+=1
```

```python
qs_the_arwu_df_inner_merge.index
```

```python
qs_the_arwu_df_inner_merge
```

```python
final_rankings = qs_the_arwu_df_inner_merge.drop(axis = 1, columns = ['Country_x', 'Country_y'])
```
